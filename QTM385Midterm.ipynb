````python
# Your answers here

class Matrix:
    """
    Matrix Class
    
    Great to do Linear Algebra!
    """
    # Constructor
    def __init__(self, num, nrow):
        """
        Constructor:
        Receives:
            num: list of numbers
            nrow: number of rows
        
        Return: 
            Matrix object
        
        """
        if len(num) % nrow != 0:
            raise ValueError('The matrix cannot be created. Vector length is not divisible by row number')

        ncol = len(num) // nrow
        self.num = num
        self.dim = [nrow, ncol]
        self.nrow = nrow
        self.ncol = ncol
        mat = []
        for i in range(0, len(num), ncol):
            mat.append(num[i:i + ncol])

        self.mat = mat
        
       
    # Print representation
    def __str__(self):
        """
        Prints the matrix when using the command print()
        """
        res = """"""
        for i in range(0, len(self.mat)):
            res += str(self.mat[i]) + '\n'
        return res


    # Representation
    def __repr__(self):
        """
        __repr__:
            Prints the object description
            In this case, the matrix
        """
        return "Matrix dimension {dim}".format(dim = self.dim)

    
    # Test to see if square matrix
    def is_square(self):
        """
        is_square:
            Receives:
                Matrix
                
            Return:
                True if the matrix is square
                False otherwise
        """
        return self.nrow == self.ncol

            
    # Addition (see https://en.wikipedia.org/wiki/Matrix_addition)
    def __add__(self, other):
        """
        __add__:
            Implements the sum of matrices
            
            Receives:
                Two matrices
                
            Returns
                Subtraction of matrices as a Matrix or error if not conform
        """
        
        res = []
        if self.nrow != other.nrow or self.ncol != other.ncol:
            raise ValueError('Error. The matrices do not conform.')
        else:
            for i in range(len(self.mat)):
                res.append(self.mat[i] + other.mat[i])
        my_matrix = Matrix([0]*len(res)**2, len(res))
        my_matrix.mat = res
        return my_matrix

    # Subtraction
    def __sub__(self, other):
        """
        __sub__:
            Implements the subtraction of two matrices
            
            Receives:
                Two matrices
                
            Returns:
                Subtraction of matrices as a Matrix or error if not conform
        """
        res = []
        if self.nrow != other.nrow or self.ncol != other.ncol:
            raise ValueError('Error. The matrices do not conform.')
        else:
            for i in range(len(self.num)):
                res.append(self.num[i] - other.num[i])

        new = []
        for i in range(0, len(res), self.ncol):
            new.append(res[i:i + self.ncol])
    
        my_matrix = Matrix([0]*len(new)**2, len(new))
        my_matrix.mat = new
        return my_matrix

        
    # Scalar multiplication (https://en.wikipedia.org/wiki/Scalar_multiplication)
    def __mul__(self, num):
        """
        __mul__:
            Implements scalar multiplication
            
            Receives:
                Matrix
                float or int
                
            Returns:
                Matrix
        """

        new = []
        for i in range(0, self.nrow):
            for j in range(0, self.ncol):
                new.append(self.mat[i][j] * num)

        my_mat = Matrix([0]*len(new), self.nrow)
        my_mat.mat = new
        return Matrix(my_mat.mat, self.nrow)
        
    # Extract row
    def exrow(self, i):
        """
        exrow:
            Extracts row of a Matrix
            
            Receives:
                Matrix
                Index (int)
                
            Returns:
                List with row or error if index out of bounds
        """
        self.i=i

        if self.i > self.nrow:
            raise ValueError('Error. Invalid row.')
        else:
            return self.mat[self.i]
    
    # Extract column
    def excol(self, i):
        """
        exrow:
            Extracts column of a Matrix
            
            Receives:
                Matrix
                Index (int)
                
            Returns:
                List with column or error if index out of bounds
        """
        self.i= i
        if self.i > self.ncol:
            raise ValueError('Error. Invalid column.')
        else:
            c = [self.mat[j][i] for j in range(len(self.mat))]
            return c

    
    # Matrix multiplication (see https://en.wikipedia.org/wiki/Matrix_multiplication)
    def __pow__(self, other):
        """
        __pow__:
            Implements the product of two matrices
            
            Receives:
                Two Matrices
                
            Returns:
                Matrix or error if not conform
        """
        if other.nrow != self.ncol:
            raise ValueError('Matrices do not conform')
        else:
            rows, cols = self.nrow, other.ncol
            # Create empty matrix
            res = []
            
            # iterate through rows of self
            for i in range(len(self.mat)):
                myrow = self.exrow(i)
                # iterate through columns of other
                for j in range(len(other.mat[0])):
                    mycol = other.excol(j)
                    # iterate through rows of other
                    aux = 0
                    for k in range(len(mycol)):
                        aux += mycol[k]*myrow[k]
                    res.append(aux)

        my_matrix = Matrix(res, self.nrow)
        return my_matrix
    
    # Extract submatrix for minor (see https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def subm(self, row, col):
        """
        subm:
            Extracts a submatrix for minor and cofactor computations
            
            Receives:
                Matrix
                row (int)
                col (int)
                
            Returns:
                Matrix
        """
        new = []
        for i in range(0, len(self.num), self.ncol):
            new.append(self.num[i:i + self.ncol])
        X = new
        del(X[row]) # delete the row
        for n in range(len(X)):
            del(X[n][col])  # delete the column elements of the rows
        
        my_matrix = Matrix([0]*len(X)**2, len(X))
        my_matrix.mat = X
        return my_matrix
    
    # Determinant (see https://en.wikipedia.org/wiki/Determinant)
    def det(self):
        """
        det:
            Compute the determinant of a matrix by cofactor expansion
            
            Receives:
                Matrix
                
            Returns:
                float, int or an error if matrix is not square.
        """
        
        if self.is_square() == False:
            raise ValueError('Determinant cannot be calculated on a non-square matrix.')
        else:
            if self.nrow == 2 & self.ncol == 2:
                return self.mat[0][0]*self.mat[1][1]-self.mat[1][0]*self.mat[0][1]
            else:
                s = 0 
                for i in range(0, self.ncol):
                    m = self.subm(0,i)
                    s = s + ((-1)**i) * self.mat[0][i] * m.det()
                return s
            
    # Transpose
    def t(self):
        """
        t:
            Computes the transpose of a matrix
            
            Receives:
                Matrix
                
            Returns:
                Matrix
        """
        # Section 1: Get matrix dimensions
        trows = len(self.mat)
        tcols = len(self.mat[0])
        
        # Section 2: Create empty transposed matrix
        M = []
        for i in range(tcols):
            M.append(self.excol(i))
        my_matrix = Matrix([0]*trows*tcols, tcols)
        my_matrix.mat = M
        return my_matrix

    
    # Cofactor matrix (see https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def cofm(self):
        """
        cofm:
            Computes the Cofactor Matrix
            
            Receives:
                Matrix
                
            Returns:
                Matrix or error if matrix not square.
        """
        if self.nrow != self.ncol:
            raise ValueError('Matrix is not square')
        else:
            determinant = self.det()
            #special case for 2x2 matrix:
            if self.nrow == 2:
                return [[m[1][1]/determinant, 
                         -1*m[0][1]/determinant],
                        [-1*m[1][0]/determinant, 
                         m[0][0]/determinant]]

            #find matrix of cofactors
            cofactors = []
            for i in range(len(self.mat)):
                cofactorRow = []
                for j in range(len(self.mat)):
                    minor = self.subm(i,j)
                    cofactorRow.append(((-1)**(i+j)) * minor.det())
                cofactors.append(cofactorRow)

            my_matrix = Matrix([0]*len(cofactors)**2, len(cofactors))
            my_matrix.mat = cofactors

        return my_matrix

    
    # Inverse matrix (see the inverse definition after buiding the cofactor matrix: 
    #                 https://en.wikipedia.org/wiki/Minor_(linear_algebra))
    def inv(self):
        """
        inv:
            Computes Inverse Matrix using cofactor matrix
            
            Receives:
                Matrix
                
            Returns:
                Matrix or error if matrix is singular or not square.
        """

        if self.is_square()==False:
            return ValueError('Error: Matrix is not square. Inverse cannot be computed')
        elif self.det()==0:
            return ValueError('Error: Determinant = 0')
        else:
            
            return self.cofm().t() * (1/self.det())

# Running Regression (Problem 11):
X = Matrix([1, -4, 0, 1, -3, 0, 1, -2, 0, 1, 
            -1, 0, 1, 0, 0, 1, 1, 1, 1, 2, 
            1, 1, 3, 1, 1, 4, 1, 1, 5, 1], 10)

y = Matrix([-3.56518269407931, -2.96460021969053, -1.48659413971178, 
            0.700322671620325, 0.902347734124384, 1.29721825127923, 
            1.92302031156774, 3.88376671356276, 2.23958110933972, 6.5610333044646], 10)

print((X.t() ** X).inv()**X.t()**y)
````
